**3. Modeling**

Consider creating a function that will calculate a ratio of house prices, comparing the price of a house in 2018 to the price in 1998.

Consider calling this function create_price_ratio.

You'd want this function to:
1. Take a filter of dfg, specifically where this filter constrains the London_Borough, as an argument. For example, one admissible argument should be: dfg[dfg['London_Borough']=='Camden'].
2. Get the Average Price for that Borough, for the years 1998 and 2018.
4. Calculate the ratio of the Average Price for 1998 divided by the Average Price for 2018.
5. Return that ratio.

Once you've written this function, you ultimately want to use it to iterate through all the unique London_Boroughs and work out the ratio capturing the difference of house prices between 1998 and 2018.

Bear in mind: you don't have to write a function like this if you don't want to. If you can solve the brief otherwise, then great! 

***Hint***: This section should test the skills you acquired in:
- Python Data Science Toolbox - Part One, all modules

```python
def create_price_ratio(arg):
    '''
    Take dfg as argument, calculate the ratio for each borough
    returns a dictionary contains each borough with its growth ratio
    '''
    ratio_dict = {}
    for i in arg['London_Borough'].unique():
        borough_series = arg[arg['London_Borough'] == i]      
        y1998 = float(borough_series[borough_series['Year']==1998]['AveragePrice'])
        y2018 = float(borough_series[borough_series['Year']==2018]['AveragePrice'])
        ratio = [y2018/y1998]
        ratio_dict[i]=ratio
    return ratio_dict
```

```python
#check the dict
change_1998_2018 = create_price_ratio(dfg)
change_1998_2018
```

```python
#transform into dataframe
change_1998_2018 = pd.DataFrame(change_1998_2018)
change_1998_2018
```

```python
#transpose it and reset index
change_1998_2018_T = change_1998_2018.transpose()
change_1998_2018_T = change_1998_2018_T.reset_index()
change_1998_2018_T
```

```python
# reset column names
change_1998_2018_T.rename(columns={'index':'Borough',0:'Ratio'}, inplace=True)
change_1998_2018_T
```

```python
#sort the value from higher to lower
# Hackney has the highest ratio
change_1998_2018_T.sort_values(by = 'Ratio',ascending=False)
```
